#4.Using the given x and y data,generate a 3d surface plot to visualize the function z=sin(sqrt{x^{2}+y^{2}})

```python
import numpy as np
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))
data={'X':x.flatten(),'Y':y.flatten(),'Z':z.flatten()}
df=pd.DataFrame(data)
fig=go.Figure(data=[go.Surface(x=df['X'].values.reshape(100,100),y=df['Y'].values.reshape(100,100),z=df['Z'].values.reshape(100,100))])
fig.update_layout(title='3D Surface Plot',autosize=False,width=800,height=600,margin=dict(l=65,r=50,b=65,t=90))
fig.show()
```

#2.Create a bokeh scatter plot using randomly generated x and y values.Use different sizes and colors for the markers based on the 'size' and 'color' columns.

```python
from bokeh.models import ColumnDataSource

np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
size = np.random.randint(5, 50, size=n)
color = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)

data = pd.DataFrame({'x': x, 'y': y, 'size': size, 'color': color})

source = ColumnDataSource(data)

p = figure(title="Bokeh Scatter Plot", x_axis_label='X Axis', y_axis_label='Y Axis')
p.scatter('x', 'y', size='size', color='color', source=source, legend_field='color', fill_alpha=0.6)

show(p)
```

#3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits=['Apples','Oranges','Bananas','Pears']
counts=[20,25,30,35]

data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits", line_color='white', fill_color='blue')

show(p)
```

#4.Create a bokeh histogram to visualize the distribution of given data.

```python
data_hist=np.random.randn(1000)
hist ,edges=np.histogram(data_hist,bins=30)
p = figure(title='Histogram of Random Data',
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools='')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.5)
show(p)
```

#5.Create a bokeh heatmap using the provided dataset.

```python
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

data_heatmap=np.random.rand(10,10)
x=np.linspace(0,1,10)
y=np.linspace(0,1,10)
xx,yy=np.meshgrid(x,y)

data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}
p = figure(title="Heatmap Example", x_axis_label='X-axis', y_axis_label='Y-axis',
           tools="", width=600, height=600)

mapper = linear_cmap(field_name='value', palette=Viridis256, low=data['value'].min(), high=data['value'].max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=data, line_color=None, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
