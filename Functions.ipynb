#6.Explain the concept of generators in Python and how they are defined.
  Generator function acts like iterator so instead storing values in memory it makes it iterable using a yield statement. That's why it is fast also.

```python
#Understand generator function with fibonacci series.
def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a,b=b,a+b
```

#10.What is the difference between map(), reduce() and filter() functions in Python?

```python
#Map Function
l=["1","2","3"]
list(map(lambda x:int(x),l))


```

#1.Write a python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(n):
  sum=0
  for i in n:
    if i%2==0:
      sum=sum+i
  return sum
```

#4.Write a Python function that checks if a given number is prime or not from 1 to 100.

```python
def checkprime(num):
  if num < 2:
    return 0
  else:
    x=num//2
    for j in range(2,x + 1):
      if num%j==0:
        return 0
    return 1

for i in range(1,100+1):
  if checkprime(i):
    print(i,end=" ")
```

#6.Write a generator function in Python that yields powers of 2 upto a given exponent.

```python
def power_generator(base,exponent):
   result=1
   for i in range(exponent+1):
    yield result
    result=result*base

base=int(input("enter the base number"))
exponent=int(input("enter the exponent number"))

power_gen=power_generator(base,exponent)
print(f"powers of{base} upto a exponent {exponent}:")
for power in power_gen:
    print(power)
```

#7.Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
  try:
    with open(file_path, 'r') as file:
       for line in file:
         yield line.strip()
  except FileNotFoundError:
    print(f"Error:The file at {file_path} was not found.")
  except Exception as e:
    print(f"An error occured: {e}")
```

#9.Write a python that uses map() to convert a list of temperatures from celcius to fahrenheit.

```python
def celcius_to_fahrenheit(celcius):
  return (9/5)*celcius +32

celcius_temp=[0,32,45,80]

fahrenheit_temp=list(map(celcius_to_fahrenheit,celcius_temp))

print("celcius temperatures:",celcius_temp)
print("fahrenheit temperatures:",fahrenheit_temp)
```
